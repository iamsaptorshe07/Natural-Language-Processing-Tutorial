<h1>Part Of Speech </h1>

```
import spacy
import nltk
nlp = spacy.load('en_core_web_sm')
```

```
doc = nlp(u"The quick brown fox jumped over the lazy dog's back")
print(doc.text)
```

```
# Print part of speech type of each & every word
for i in doc:
  print("{} ------- {} ------ {} ------- {}".format(i.text,i.pos_,i.tag_,spacy.explain(i.tag_)))
```

<h1>Name Entity Recognizer </h1>

```
# Show the entities 
def show_ents(doc):
  if doc.ents:
    for ents in doc.ents:
      print(ents.text)
  else:
    print('no entitites found')
```

```
with open('leavelicesne.txt', encoding='unicode_escape') as file:
  docx = nlp(file.read())
  print(docx)
```

```
show_ents(docx)
```
